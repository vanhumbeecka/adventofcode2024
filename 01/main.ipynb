```python
with open("real.txt") as f:
    lines = f.readlines()
    for i in range(len(lines)):
        lines[i] = [int(x) for x in lines[i].strip().split()]

```

```python
# left vs right
left = []
right = []
for line in lines:
    left.append(line[0])
    right.append(line[1])

# # add index to left and right
# left = np.array(left)
# right = np.array(right)
# left = np.column_stack((left, np.arange(len(left))))
# right = np.column_stack((right, np.arange(len(right))))

# # sort left and right
# left = left[left[:, 0].argsort()]
# right = right[right[:, 0].argsort()]
```

```python
left.sort()
right.sort()

# left, right
```

```python

diff = 0
for i in range(len(left)):
    diff += abs(left[i] - right[i])

print(diff)
```

```python
print(left)
np.searchsorted(left, 3, side='left'), np.searchsorted(left,3, side='right')
```

```python
# binary search first occurence of right[i] in left
simm = 0
for i in range(len(left)):
    count = np.searchsorted(right, left[i], side='right') - np.searchsorted(right, left[i], side='left')
    simm += left[i] * count

print(simm)
```
