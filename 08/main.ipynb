```python
lines = []
antennas = dict()

with open('input.txt') as f:
    h = 0
    for line in f:
        lines.append(line.strip())
        width = len(line.strip())
        chars = list(line.strip())
        for idx, char in enumerate(chars):
            if char != '.':
                if antennas.get(char) is None:
                    antennas[char] = []
                antennas[char].append((idx, h))

        h += 1
    height = len(lines)

print(f'width: {width}, height: {height}')
print(antennas)
```

```python
import itertools

# Generate permutations of pairs of list items
pair_permutations = {key: list(itertools.permutations(value, 2)) for key, value in antennas.items()}
print(pair_permutations)
```

```python
print(list(sorted(antennas.keys())))
```

```python
vectors = {key: [(pair[1][0] - pair[0][0], pair[1][1] - pair[0][1]) for pair in pairs] for key, pairs in pair_permutations.items()}
print(vectors)
```

```python
coordinates = {key: [(pair[1][0] + vector[0], pair[1][1] + vector[1]) for pair, vector in zip(pair_permutations[key], vectors[key])] for key in vectors.keys()}
print(coordinates)
```

```python


coords = []
for key in coordinates.keys():
    for coordinate in coordinates[key]:
        if coordinate[0] < 0 or coordinate[0] >= width or coordinate[1] < 0 or coordinate[1] >= height:
            continue
        if coordinate not in coords:
            coords.append(coordinate)
        # print(f'{key} can see {coordinate}')
        
print(len(coords))
```

```python
# Create an empty grid
map_grid = [['.' for _ in range(width)] for _ in range(height)]
print(map_grid)

# Mark the coordinates on the grid
for coord in coords:
    print(coord)
    map_grid[coord[1]][coord[0]] = '#'

# Print the grid
for row in map_grid:
    print(''.join(row))
```
