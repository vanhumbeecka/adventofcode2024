```python
with open('input.txt') as f:
    line = f.readline().strip()

print(line)
```

```python
current_pos = 0
disk_map = dict()
empty_idx_list: list[int] = []
file_idx_list: list[int] = []
empty_idx_size_dict = dict()
file_idx_size_dict = dict()

for idx, char in enumerate(list(line)):
    block_len = int(char)
    file_id = 0
    count_empty = 0
    count_full = 0
    for i in range(block_len):
        if idx % 2 == 0:
            count_full += 1
            file_id = idx // 2
            file_idx_list.append(current_pos+i)
        else:
            count_empty += 1
            file_id = None
            empty_idx_list.append(current_pos+i)
        disk_map[current_pos+i] = file_id
    if idx % 2 == 1:
        empty_idx_size_dict[current_pos] = count_empty
    else:
        file_idx_size_dict[current_pos] = count_full
    current_pos += block_len
    

def print_disk_map(disk_map):
    for key in disk_map.keys():
        if disk_map[key] is None:
            print('.', end='')
        else:
            print(f"{disk_map[key]}", end='')

print_disk_map(disk_map)
print()
print("File idx list:")
print(file_idx_list)
print()
print("Empty idx list:")
print(empty_idx_list)
print()
print("Empty idx size dict:")
print(empty_idx_size_dict)
print()
print("File idx size dict:")
print(file_idx_size_dict)

```

```python
from tqdm import tqdm

disk_map_copy = disk_map.copy()
for i, file_idx in tqdm(enumerate(reversed(file_idx_list))):
    if empty_idx_list[i] > file_idx:
        break
    disk_map_copy[empty_idx_list[i]] = disk_map[file_idx]
    disk_map_copy[file_idx] = None
    
```

```python
total = 0
for key, value in disk_map_copy.items():
    if value is None:
        continue
    else:
        total += key * value

print(total)
```
