```python

with open('input.txt', 'r') as file:
    lines = file.readlines()

data = []
for line in lines:
    parts = line.strip().split()
    p = tuple(map(int, parts[0][2:].split(',')))
    v = tuple(map(int, parts[1][2:].split(',')))
    data.append((p, v))

print(data)
```

```python
HEIGHT = 103
WIDTH = 101
STEPS = 100

# HEIGHT = 7
# WIDTH = 11

def move_steps(robots, steps, width, height):
    pos = []
    for robot in robots:
        p, v = robot
        p_new = ((p[0] + v[0] * steps) % width, (p[1] + v[1] * steps) % height)
        pos.append(p_new)
    return pos

pos = move_steps(data, STEPS, WIDTH, HEIGHT)
# for robot in data:
#     p, v = robot
#     p_new = ((p[0] + v[0] * STEPS) % WIDTH, (p[1] + v[1] * STEPS) % HEIGHT)
#     pos.append(p_new)

print(pos)
```

```python
def quadrant(pos: list[tuple[int, int]], width: int, height: int) -> int:
    q1, q2, q3, q4 = 0, 0, 0, 0
    for p in pos:
        if p[0] < (width-1) / 2 and p[1] < (height-1) / 2:
            q1 += 1
        elif p[0] >= (width+1) / 2 and p[1] < (height-1) / 2:
            q2 += 1
        elif p[0] < (width-1) / 2 and p[1] >= (height+1) / 2:
            q3 += 1
        elif p[0] >= (width+1) / 2 and p[1] >= (height+1) / 2:
            q4 += 1
        
    print(q1, q2, q3, q4)
    return q1 * q2 * q3 * q4

print(quadrant(pos, WIDTH, HEIGHT))
```

```python
def print_pos(pos: list[tuple[int, int]], width: int, height: int):
    for y in range(height):
        for x in range(width):
            if (x, y) in pos:
                print('#', end='')
            else:
                print('.', end='')
        print()


pos = move_steps(data, 11, WIDTH, HEIGHT)
print_pos(pos, WIDTH, HEIGHT)

```
