```python
import numpy as np
import re
```

```python
with open('input.txt') as f:
    lines = f.readlines()
print(lines)
# concat lines from list
all_lines = ''.join(lines)
```

```python
print(all_lines)
```

```python
from typing import Optional

def find_matches(text: str):
    # regex to find mul(X,Y) in string
    mul = re.compile(r'mul\((\d+),(\d+)\)')

    # find all matches
    return mul.findall(text)

def find_matches_from_start(text: str) -> Optional[re.match]:
    # regex to find mul(X,Y) in string
    mul = re.compile(r'^mul\((\d+),(\d+)\)')

    # find all matches
    return mul.match(text)
```

```python
print(find_matches("mul(4*"))
print(find_matches("mul(6,9!"))
print(find_matches("?(12,34)"))
print(find_matches("mul ( 2 , 4 )"))
print(find_matches_from_start("mul(2,4)mul(3,5)"))
print(find_matches_from_start("mul(2,a4)mul(a3,5)"))
```

```python
matches = find_matches(all_lines)

total = 0
for match in matches:
    # convert to integers
    x, y = map(int, match)

    # multiply x and y
    result = x * y

    # replace mul(X,Y) with result
    total += result

print(total)
    
```

```python
result = 0
enabled = True
for idx in range(len(all_lines)):
    # first check if "do()" or "don't()" is in the beginning of the line
    if all_lines[idx:].startswith('do()'):
        print("do() found")
        enabled = True
    elif all_lines[idx:].startswith("don't()"):
        print("don't() found")
        enabled = False
    
    if not enabled:
        continue

    match = find_matches_from_start(all_lines[idx:])
    if match:
        print(match.groups())
        x, y = map(int, match.groups())
        result += x * y if enabled else 0

print(result)
```
